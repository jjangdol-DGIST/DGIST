### 실습 문제 1. 주어진 신호에 대해서 정제할 수 있는 함수를 적절히 적용하시오.
주어진 함수 'bw_lowpass'와 'bw_highpass'만을 사용해 이 신호를 정제해 보세요.
가장 마지막 셀에 있는 result에 적절한 함수를 적용한 결과를 넣어보세요.

```python
import numpy as np
import scipy
import scipy.signal
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 1, 500, endpoint=False)
signal = np.sin(2 * np.pi * 10 * t) + 0.5 * np.random.randn(500)
plt.plot(t, signal)
plt.show()
```

```python
def bw_lowpass(signal):
    filter_length = 51  
    cutoff_frequency = 0.1  # cutoff frequency
    n = np.arange(filter_length)
    h = np.sinc(2 * cutoff_frequency * (n - (filter_length - 1) / 2))
    h = h / np.sum(h)
    filtered = np.convolve(signal, h, mode='same')
    return filtered

def bw_highpass(signal):
    filter_length = 51  
    cutoff_frequency = 0.1  # cutoff frequency
    n = np.arange(filter_length)
    h = np.sinc(2 * cutoff_frequency * (n - (filter_length - 1) / 2))
    h = h * (-1) ** n 
    h = h / np.sum(h)
    filtered = np.convolve(signal, h, mode='same')
    return filtered
```

```python
### your code here
result = bw_lowpass(signal)
plt.plot(t, result)
plt.show()
```
