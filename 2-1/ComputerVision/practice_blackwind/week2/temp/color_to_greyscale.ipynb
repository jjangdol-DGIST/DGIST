```python
import numpy as np
from PIL import Image
```

```python
img_name = "sample.bmp" # input("파일 이름을 입력하세요: ")
```

```python
img = Image.open(img_name)
img_array = np.array(img, dtype=np.uint64)
img
```

```python
import matplotlib.pyplot as plt
```

```python
row, col, _ = img_array.shape
print(img_array.shape)
```

```python
# 예제 1: color image를 lightness 기법을 사용해 greyscale 이미지로 바꾸시오.
result_image = np.zeros((row, col))
#### your code here
for i in range(row):
    for j in range(col):
        result_image[i][j] = int((max(img_array[i][j][0], img_array[i][j][1], img_array[i][j][2])+min(img_array[i][j][0], img_array[i][j][1], img_array[i][j][2]))/2)
#### your code here
lightness_array = result_image.astype(np.uint8)
lightness = Image.fromarray(lightness_array)
```

```python
lightness
```

```python
# 문제 1: color image를 average 기법을 사용해 greyscale 이미지로 바꾸시오.
result_image = np.zeros((row, col))
### your code here
for i in range(row):
    for j in range(col):
        result_image[i][j] = int(((img_array[i][j][0] + img_array[i][j][1] + img_array[i][j][2]) / 3 ))
### your code here
average_array = result_image.astype(np.uint8)
average = Image.fromarray(average_array)
```

```python
average
```

```python
# 문제 2: color image를 lumiosity 기법을 사용해 greyscale 이미지로 바꾸시오.
result_image = np.zeros((row, col))
### your code here
for i in range(row):
    for j in range(col):
        result_image[i][j] = int(((0.30 * img_array[i][j][0] + 0.59 * img_array[i][j][1] + 0.11 * img_array[i][j][2])))
### your code here
lumio_array = result_image.astype(np.uint8)
lumio = Image.fromarray(lumio_array)
```
