```python
import numpy as np
from PIL import Image
```

```python
img_name = "sample.bmp" # input("파일 이름을 입력하세요: ")
```

```python
img = Image.open(img_name)
img
```

```python
img_array = np.array(img)
```

```python
img_array.shape
```

```python
img_0 = Image.fromarray(img_array[:,:,0]) # red
img_1 = Image.fromarray(img_array[:,:,1]) # green
img_2 = Image.fromarray(img_array[:,:,2]) # blue
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img_0, cmap = 'gray')
plt.show()
```

```python
plt.imshow(img_1, cmap = 'gray')
plt.show()
```

```python
plt.imshow(img_2, cmap = 'gray')
plt.show()
```
