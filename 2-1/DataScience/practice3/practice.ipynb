```python
df = pd.read_csv('dataset.csv')

df.replace("", np.nan, inplace=True)
```

```python
df = df.iloc[4:]

df.dropna(how='all', inplace=True)

new_columns = ['id', 'name', 'district', 'address', 'latitude', 'longitude', 'date', 'amount_LCD', 'amount_QR', 'type']

df.columns = new_columns

df.set_index('id', inplace=True)

# df.index = df.index.astype('int')
```

```python

df['amount'] = df.apply(lambda row: row['amount_LCD'] if not pd.isna(row['amount_LCD']) else row['amount_QR'], axis=1)

df.drop(['amount_LCD', 'amount_QR'], axis=1, inplace=True)

df.index = df.index.astype('int')
```

```python
year_counts = df['date'].str.extract(r'(\d{4})', expand=False).value_counts().sort_index()

for year, count in year_counts.iteritems():
    print(f'{year}: {count}')
```
